# SMT Research — SPY/QQQ visual inspection
Plot price with detected swing points and SMT divergence events.

```python
import sys, os
sys.path.insert(0, os.path.abspath('..'))
import matplotlib.pyplot as plt
from data import YFinanceLoader, align
from core import detect_swings, detect_smt

loader = YFinanceLoader()
a = loader.fetch('SPY', '2023-01-01', '2023-12-31')
b = loader.fetch('QQQ', '2023-01-01', '2023-12-31')
a, b = align(a, b)
events = detect_smt(a, b, window=5)
print(f'Found {len(events)} SMT events')
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 8), sharex=True)
ax1.plot(a.index, a['close'], label='SPY')
ax2.plot(b.index, b['close'], label='QQQ', color='orange')
for e in events:
    color = 'red' if e.direction == 'bearish' else 'green'
    ax1.axvline(e.timestamp, color=color, alpha=0.3)
    ax2.axvline(e.timestamp, color=color, alpha=0.3)
ax1.legend(); ax2.legend(); plt.tight_layout(); plt.show()
```
